```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')
```

## Step 1: Import Libraries

Import all the libraries we need:
- **numpy**: For numerical operations
- **sklearn**: Machine learning algorithms and tools
- **warnings**: To suppress warning messages

# Your First MLflow Experiment

This notebook demonstrates the basics of MLflow experiment tracking.

**What you'll learn:**
- How to create a simple classification model
- How to track model parameters
- How to log performance metrics
- How to save models with MLflow

Perfect for beginners getting started with MLflow!

```python
# Step 1: Create an imbalanced binary classification dataset
X, y = make_classification(n_samples=1000, n_features=10, n_informative=2, n_redundant=8, 
                           weights=[0.9, 0.1], flip_y=0, random_state=42)

np.unique(y, return_counts=True)
```

## Step 2: Create a Dataset

Generate a synthetic dataset for binary classification:
- **n_samples=1000**: Create 1000 data points
- **n_features=10**: Each sample has 10 features
- **weights=[0.9, 0.1]**: Imbalanced classes (90% class 0, 10% class 1)
- This simulates real scenarios like fraud detection or spam classification

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

## Step 3: Split Data into Train and Test Sets

Divide the dataset:
- **70% for training**: Model learns patterns from this data
- **30% for testing**: Model is evaluated on unseen data
- **stratify=y**: Maintains the same class ratio in both sets

```python
# Define the model hyperparameters
params = {
    "solver": "lbfgs",
    "max_iter": 1000,
    "random_state": 8888,
}

# Train the model
lr = LogisticRegression(**params)
lr.fit(X_train, y_train)

# Predict on the test set
y_pred = lr.predict(X_test)

report = classification_report(y_test, y_pred)
print(report)
```

## Step 4: Train a Logistic Regression Model

**What happens here:**
1. Define model parameters (settings)
2. Create and train the Logistic Regression model
3. Make predictions on test data
4. Generate a performance report

**Logistic Regression** is a simple algorithm good for binary classification problems.

```python
report_dict = classification_report(y_test, y_pred, output_dict=True)
report_dict
```

## Step 5: Convert Report to Dictionary

Convert the classification report to dictionary format for easier processing.
This makes it simple to extract specific metrics like accuracy, precision, and recall.

```python
import mlflow
```

## Step 6: Import MLflow

Install setuptools (if needed) and import MLflow for experiment tracking

```python
mlflow.set_experiment("First Experiment")
mlflow.set_tracking_uri(uri="http://127.0.0.1:5000/")

with mlflow.start_run():
    mlflow.log_params(params)
    mlflow.log_metrics({
        'accuracy': report_dict['accuracy'],
        'recall_class_0': report_dict['0']['recall'],
        'recall_class_1': report_dict['1']['recall'],
        'f1_score_macro': report_dict['macro avg']['f1-score']
    })
    mlflow.sklearn.log_model(lr, "Logistic Regression")  
```

## Step 7: Track Your First Experiment with MLflow!

**This is where MLflow magic happens:**

1. **Set experiment name**: Creates or uses existing "First Experiment"
2. **Set tracking URI**: Points to local MLflow server at http://127.0.0.1:5000
3. **Start a run**: Creates a new experiment record
4. **Log parameters**: Records model settings (solver, max_iter, random_state)
5. **Log metrics**: Saves performance metrics (accuracy, recall, f1-score)
6. **Log model**: Saves the trained model for later use

**After running this cell:**
- Open your browser and go to http://127.0.0.1:5000
- You'll see your experiment with all logged information!
- You can compare different runs and track your progress

Congratulations! You've just tracked your first ML experiment! 🎉
